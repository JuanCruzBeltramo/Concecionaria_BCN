1. Origen

AutoConexa nace de la necesidad de modernizar el mercado de vehículos de segunda mano. Identificando una brecha en la forma en que los compradores y vendedores interactúan, la empresa busca simplificar y agilizar este proceso. Su fundación se basa en el uso de tecnología avanzada para conectar a ambas partes de manera eficiente y transparente.

  

2. Modelo de Negocio

AutoConexa opera como una plataforma de dos caras. Por un lado, permite a los vendedores (tanto particulares como concesionarios) listar sus vehículos con información detallada y fotografías de alta calidad. Por otro lado, ofrece a los compradores una interfaz fácil de usar para buscar y comparar vehículos. La empresa genera ingresos a través de comisiones por transacción y publicidad premium para listados destacados.

  

3. Directiva y RRHH

El equipo directivo está compuesto por expertos en tecnología, marketing automotriz y gestión empresarial. La contratación de personal se centra en perfiles con experiencia en desarrollo web, análisis de datos, atención al cliente y marketing digital. Se promueve una cultura de trabajo colaborativa y orientada a resultados.

  

4. Operaciones

Las operaciones de AutoConexa se centran en mantener y mejorar la plataforma, asegurar la calidad de los anuncios, y proporcionar soporte al cliente. Esto incluye un equipo de desarrollo continuo de la plataforma, un equipo de control de calidad para verificar los anuncios, y un equipo de atención al cliente para asistir a usuarios y resolver problemas.

  

5. Mercado y Competencia

El mercado objetivo inicial son los compradores y vendedores de vehículos en España. La competencia incluye otros sitios web de compraventa de vehículos y concesionarios tradicionales. AutoConexa se diferencia por su facilidad de uso, información detallada y herramientas de comparación avanzadas.

  

6.Financiamiento

AutoConexa se estableció formalmente en septiembre de 2014, y gracias a su modelo de negocio innovador, fue seleccionada por Conector, una de las principales aceleradoras españolas, para formar parte de su programa de aceleración a principios de 2015. Esto culminó con una primera ronda de financiación en febrero de 2015, donde se recaudaron aproximadamente 140.000 euros de un grupo de business angels. La confianza en el proyecto se solidificó con la incorporación de un inversor clave, que también se convirtió en miembro fundador.

  

Tras mostrar seis meses de operaciones exitosas y métricas convincentes, AutoConexa atrajo una segunda ronda de financiación en agosto de 2015, consiguiendo unos 300.000 euros a través de The Crowd Angel, una plataforma de crowdfunding. Las operaciones robustas y el crecimiento sólido fomentaron una tercera ronda en noviembre del mismo año, liderada por Antai y apoyada por business angels de renombre, con una inversión significativa de 2 millones de euros.

  

El crecimiento continuó y, después de nueve meses y medio, la empresa cerró una cuarta ronda en agosto de 2016, esta vez liderada por Seaya Ventures, con un impresionante monto de 8 millones de euros. ### Eliminamos valores  "Outliers".

```python
numericas=['price','price_financed','kms','power','doors']

# FUNCIÓN QUE DEVUELVE LOS NÚMEROS DE FILA EN LOS QUE LA VARIABLE EN CUESTIÓN ES OULIER
def outliers(var):
    q1=var.quantile(0.25)
    q3=var.quantile(0.75)
    riq=q3-q1
    sup=q3+1.5*(riq)
    inf=q1-1.5*(riq)
    outl=(var>sup) | (var<inf)
    return outl

for i in df.loc[:,numericas]:
    print("Cantidad de oultliers de la variable",i,":",outliers(df[i]).sum())
# Eliminar outliers
# Tu código
for i in df.loc[:,numericas]:
    df=df[~outliers(df[i])]
df.sample(5)
```

```python
manual_count = df[df['shift'] == 'Manual'].shape[0]
automatic_count = df[df['shift'] == 'Automático'].shape[0]

# Calcular porcentajes
total_vehicles = manual_count + automatic_count
manual_percentage = (manual_count / total_vehicles) * 100
automatic_percentage = (automatic_count / total_vehicles) * 100

# Crear un gráfico de barras
fig, ax = plt.subplots()

# Barras
bars = plt.bar(['Manuales', 'Automáticos'], [manual_count, automatic_count], color=['blue', 'orange'])

# Añadir etiquetas
plt.xlabel('Tipo de Transmisión')
plt.ylabel('Cantidad de Vehículos')
plt.title('Comparación de Vehículos Manuales y Automáticos')

# Mostrar el porcentaje en las barras
for bar, percentage in zip(bars, [manual_percentage, automatic_percentage]):
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, f'{percentage:.1f}%', ha='center', va='bottom', color='black')

# Mostrar el gráfico
plt.show()
```

```python
numericas=['price','price_financed','kms','power','doors']

# FUNCIÓN QUE DEVUELVE LOS NÚMEROS DE FILA EN LOS QUE LA VARIABLE EN CUESTIÓN ES OULIER
def outliers(var):
    q3=var.quantile(0.75)
    q1=var.quantile(0.25)
    riq=q3-q1
    sup=q3+1.5*(riq)
    inf=q1-1.5*(riq)
    outl=(var>sup) | (var<inf)
    return outl

for i in df_for_final_model.loc[:,numericas]:
    print("Cantidad de oultliers de la variable",i,":",outliers(df_for_final_model[i]).sum())
# Eliminar outliers
# Tu código
for i in df_for_final_model.loc[:,numericas]:
    df_for_final_model=df_for_final_model[~outliers(df_for_final_model[i])]
df_for_final_model.sample(5)
```
